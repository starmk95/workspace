matplotlib.rc()Îäî MatplotlibÏùò Îü∞ÌÉÄÏûÑ ÏÑ§Ï†ï(runtime configuration)ÏùÑ Î≥ÄÍ≤ΩÌïòÎäî Ìï®ÏàòÏûÖÎãàÎã§.

```python
import matplotlib.pyplot as plt
from matplotlib import  rc

# rc('font', family = 'Malgun Gothic')
font1 = {'size':18,'color':'green'}
xdata = ['ÏïàÏßÄÏòÅ','ÌôçÏßÄÏàò','Ìô©ÏòàÎ¶∞']
ydata = [90,85,88]

plt.plot(xdata,ydata)
plt.title('ÏÑ∏ ÌïôÏÉùÏùò Íµ≠Ïñ¥ ÏÑ±Ï†Å', fontdict=font1)

plt.xlabel('Ïù¥Î¶Ñ')
plt.ylabel('ÏÑ±Ï†Å')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams.update(plt.rcParamsDefault)  # Ìè∞Ìä∏ Ï∞æÏßÄ ÏïäÎèÑÎ°ù Í∏∞Î≥∏ ÏÑ§Ï†ïÏúºÎ°ú Î≥µÏõê
date = list(range(1, 11))
min_temp = [7.3, 3.5, 7.2, 4.5, 3.5, 8.3, 7.7, 6.4, 9.3, 6.3]
max_temp = [15, 16.3, 15.2, 17.5, 18.3, 20.0, 18.8, 17.5, 14.2, 15.3]

# 2Í∞úÏùò ÏÑ† Í∑∏ÎûòÌîÑ ÏÉùÏÑ±
plt.plot(date, min_temp, label='Min Temperature', color='red', linestyle='-', marker='o')
plt.plot(date, max_temp, label='Max Temperature', color='blue', linestyle='-.', marker='x')

plt.legend(loc='best')
plt.xlabel('date')
plt.ylabel('temperature')

plt.title('Min and max temperature of ten days')

plt.show()
```
