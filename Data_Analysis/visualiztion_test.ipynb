### 📌 matplotlib.rc() 메소드란?  
matplotlib.rc()는 Matplotlib의 런타임 설정(runtime configuration)을 변경하는 함수입니다.  
그래프의 스타일, 폰트, 색상, 라인 스타일 등을 전역적으로 설정할 수 있습니다.

```python
import matplotlib.pyplot as plt
from matplotlib import  rc

# rc('font', family = 'Malgun Gothic')
font1 = {'size':18,'color':'green'}
xdata = ['안지영','홍지수','황예린']
ydata = [90,85,88]

plt.plot(xdata,ydata)
plt.title('세 학생의 국어 성적', fontdict=font1)

plt.xlabel('이름')
plt.ylabel('성적')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams.update(plt.rcParamsDefault)  # 폰트 찾지 않도록 기본 설정으로 복원
date = list(range(1, 11))
min_temp = [7.3, 3.5, 7.2, 4.5, 3.5, 8.3, 7.7, 6.4, 9.3, 6.3]
max_temp = [15, 16.3, 15.2, 17.5, 18.3, 20.0, 18.8, 17.5, 14.2, 15.3]

# 2개의 선 그래프 생성
plt.plot(date, min_temp, label='Min Temperature', color='red', linestyle='-', marker='o')
plt.plot(date, max_temp, label='Max Temperature', color='blue', linestyle='-.', marker='x')

plt.legend(loc='best')
plt.xlabel('date')
plt.ylabel('temperature')

plt.title('Min and max temperature of ten days')

plt.show()
```
